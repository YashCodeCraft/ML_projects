```python
# import packages

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
import warnings as w
w.filterwarnings('ignore')
```

```python
# extract data

melbourne_data = pd.read_csv("Melbourne_housing_FULL.csv")
melbourne_data.head(18)
```

```python
# preprocessing

melbourne_data.columns
melbourne_data = melbourne_data.drop(columns=['Longtitude', 'YearBuilt', 'Lattitude', 'Date', 'Postcode', 'Address'])
melbourne_data
```

```python
melbourne_data.isnull().sum()
```

```python
melbourne_data = melbourne_data.dropna()
melbourne_data.isnull().sum()
```

```python
melbourne_data
```

```python
melbourne_data = pd.get_dummies(melbourne_data, dtype=int)
melbourne_data
```

```python
#trainning input and output

input_dat = melbourne_data.drop(columns=['Price'])
output_dat = melbourne_data['Price']
input_train, input_test, output_train, output_test = train_test_split(input_dat, output_dat, train_size = 0.75)
```

```python
#model creating

model = LinearRegression()
model.fit(input_train, output_train)
prediction = model.predict(input_test)
print(prediction)
```

```python
model.score(input_train, output_train)
```

```python
input_train, input_test, output_train, output_test = train_test_split(input_dat, output_dat, train_size = 0.7)
model1 = LinearRegression()
model1.fit(input_train, output_train)
prediction = model1.predict(input_test)
print(model1.score(input_train, output_train))
print("overfitted",model1.score(input_test, output_test))
```

```python
# use lasso

reg2 = Lasso(alpha=50, max_iter=100, tol=0.1)
reg2.fit(input_train, output_train)
print(reg2.score(input_train, output_train))
print(reg2.score(input_test, output_test))
```

```python
# use Ridge

reg3 = Ridge(alpha=50, max_iter=100, tol=0.1)
reg3.fit(input_train, output_train)
print(reg3.score(input_train, output_train))
print(reg3.score(input_test, output_test))
```

```python
# import packages

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# extract data

places_data = pd.read_csv("C:\\To Read\\Data_sets\\Top Indian Places to Visit.csv")
places_data
```

```python
# preprocessing

places_data = places_data.drop(columns=['Unnamed: 0', "Name", 'Zone', 'State', 'City', 'time needed to visit in hrs','Establishment Year','Airport with 50km Radius','Weekly Off','Significance','DSLR Allowed','Best Time to visit'])
places_data
```

```python
places_data.isnull().sum()
```

```python
places_data = places_data[(places_data!=0).all(axis = 1)]
```

```python
places_data.isnull().sum()
```

```python
places_data
```

```python
places_data = pd.get_dummies(places_data, dtype=int)
places_data
```

```python
input_dat = places_data.drop(columns=["Entrance Fee in INR"])
output_dat = places_data['Entrance Fee in INR']
```

```python
input_train, input_test, output_train, output_test = train_test_split(input_dat, output_dat, test_size=0.25)
model = LinearRegression()
model.fit(input_train, output_train)
print(model.predict(input_test))
```

```python
score = (model.score(input_train, output_train))*100
score = int(score)
print(f"The score will be {score}%")
```
